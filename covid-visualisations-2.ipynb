```python
import datetime
import numpy as np

z = np.random.poisson(size=(len(df1['State']), len(df1['Last Updated'])))

fig = go.Figure(data=go.Heatmap(
        z=z,
        x=df1['Last Updated'],
        y=df1['State'],
        colorscale='Viridis'))

fig.update_layout(
     xaxis_nticks=36, title_text='<b>Heatmap of Poisson\'s distribution (Date vs State)<b>')


fig.show()
```
